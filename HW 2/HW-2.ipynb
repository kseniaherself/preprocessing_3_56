First I transfer my data from _csv_. 

```python
con = sqlite3.connect("mac.db")
cur = con.cursor()
cur.execute("DROP TABLE IF EXISTS detailed")
cur.execute("CREATE TABLE  (word_ortho, word_q1, onset_q1, coda_q1);")  # all the columns names

with open('lm.csv','r', encoding='utf-8') as fin:
    dr = csv.DictReader(fin)  # a comma is the default separator
    # put the columns' names again
    to_db = [(i['word_ortho'], i['word_q1'], i['onset_q1'], i['coda_q1']) for i in dr]  

# and again, the names of the columns 
cur.executemany("INSERT INTO detailed (word_ortho, word_q1, onset_q1, coda_q1) VALUES (?, ?, ?, ?);", to_db)  

for i in cur.execute("SELECT * FROM detailed"):
    print(i)
    
con.commit()
con.close()
```
